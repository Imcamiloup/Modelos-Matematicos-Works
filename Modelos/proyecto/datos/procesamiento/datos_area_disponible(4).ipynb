```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargar datos actualizados (ajustar ruta)
df = pd.read_csv("./datos_territorio/datos_area_protegida.csv", decimal=".")
```

```python
df.head(10)
```

```python
df["Area disponible"] = (163600 - df["Huella Urbana"]).round(2)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df["año"], df["Area disponible"],
         label="Area disponible",
         color="blue", 
         marker="o")

plt.title("Área Disponible")
plt.xlabel("Año")
plt.ylabel("Área Disponible (hectáreas)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
```

```python
# Guardar
df.to_csv("./datos_territorio/datos_final.csv", index=False, encoding="utf-8-sig")
```
