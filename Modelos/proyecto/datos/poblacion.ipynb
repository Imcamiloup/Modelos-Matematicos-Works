```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
# Cargar datos (ajustar ruta)
df = pd.read_csv("./datos_poblacion.csv", decimal=".")
```

```python
df.head(5)
```

```python
plt.figure(figsize=(12, 6))
# Graficar datos imputados (línea continua)
# plt.plot(
#     df["año"], 
#     df["población"], 
#     color="orange", 
#     linestyle="--", 
#     label=""
# )

# Graficar datos originales (puntos en rojo)
plt.scatter(
    df["año"],
    df["población"],
    color="orange", 
    marker="o", 
    s=10, 
    label="Datos originales"
)
plt.plot(df["año"], df["población"], color='orange', linestyle='dashed', linewidth=0.8)

# Personalizar
plt.title("Población de Bogotá (1980-2024)")
plt.xlabel("Año")
plt.ylabel("Población")
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend(loc="upper left")

# Mostrar
plt.show()
```

```python
plt.figure(figsize=(12, 6))
# Graficar datos imputados (línea continua)
# plt.plot(
#     df["año"], 
#     df["población"], 
#     color="orange", 
#     linestyle="--", 
#     label=""
# )

# Graficar datos originales (puntos en rojo)
plt.scatter(
    df["año"],
    df["natalidad"],
    color="red", 
    marker="o", 
    s=10, 
    label="Tasa Natalidad"
)
plt.plot(df["año"], df["natalidad"], color='red', linestyle='dashed', linewidth=0.8)

plt.scatter(
    df["año"],
    df["mortalidad"],
    color="green", 
    marker="o", 
    s=10, 
    label="Tasa Mortalidad"
)
plt.plot(df["año"], df["mortalidad"], color='green', linestyle='dashed', linewidth=0.8)

plt.scatter(
    df["año"],
    df["migración"],
    color="blue", 
    marker="o", 
    s=10, 
    label="Tasa migración"
)
plt.plot(df["año"], df["migración"], color='blue', linestyle='dashed', linewidth=0.8)

# Personalizar
plt.title("Tasas Natalidad/Mortalidad/Migración de Bogotá (1980-2024)")
plt.xlabel("Año")
plt.ylabel("Tasa")
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend(loc="upper left")

# Mostrar
plt.show()
```
